```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
output_data_file = 'cities_clean.csv'
cities_df = pd.read_csv(output_data_file)
cities_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude in locations
locations = cities_df[["Latitude", "Longitude"]]
humidity = cities_df['Humidity']
```

```python
# Plot Heatmap
fig = gmaps.figure(center=[37,-95],zoom_level=4)

# Create heat layer
heat_layer = gmaps.heatmap_layer(
    locations, 
    weights=humidity, 
    dissipating=False, 
    max_intensity=10,
    point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# I set temperatures in metric units so ideal temperature is between 21.1C and 26.7C, windspeed< 16.1 km/hr   
cities_ideal_df = cities_df.loc[(cities_df['Max Temp'] < 26.7) 
              & (cities_df['Max Temp'] > 21.1) 
              & (cities_df['Wind Speed'] < 16.1) 
              & (cities_df['Cloudiness'] == 0)
              & (cities_df['Humidity'] < 70)
              & (cities_df['Humidity'] >= 35)]

cities_ideal_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Create an API search for 'lodging' (closest to hotels) within 5000m of our city. 
target_type = "lodging"
radius = 5000
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
hotel_df = pd.DataFrame(data = cities_ideal_df)
# add in blank columns to our dataframe which will contain hotel co-ordinates
hotel_df['Hotel Name'] = ""
hotel_df['Lat'] = ""
hotel_df['Lng'] = ""
hotel_df.reset_index(inplace=True)
hotel_df.head()
```

```python
# Build URL using the Google Maps API
for index, row in hotel_df.iterrows():
    lat = row["Latitude"]
    lng = row["Longitude"]
    params = {
            "location": f"{lat},{lng}",
            "types": target_type,
            "radius": radius,
            "key": gkey
            }
    # call the API and pull the hotel name and it lat and long coordinates
    response = requests.get(base_url, params).json()
    results = response['results']
    print(f"Closest hotel is {results[0]['name']}")
    hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
    hotel_df.loc[index, 'Lat'] = results[0]['geometry']['location']['lat']
    hotel_df.loc[index, 'Lng'] = results[0]['geometry']['location']['lng']
```

```python
# display the complete hotel df.
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[['Lat','Lng']].astype(float)
locations_tup = list(locations.to_records(index=False))
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations_tup, info_box_content= hotel_info)
fig.add_layer(markers)

# Display figure
fig
```
